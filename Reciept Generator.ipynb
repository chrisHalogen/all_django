```python
from reportlab.lib.pagesizes import letter, A4
from reportlab.lib.units import inch

# Create's a canvas
c = canvas.Canvas('myfile.pdf')

# move the origin up and to the left
c.translate(inch,inch)
# define a large font
c.setFont("Helvetica", 14)
# choose some colors
c.setStrokeColorRGB(0.2,0.5,0.3)
c.setFillColorRGB(1,0,1)
# draw some lines
c.line(0,0,0,1.7*inch)
c.line(0,0,1*inch,0)
# draw a rectangle
c.rect(0.2*inch,0.2*inch,1*inch,1.5*inch, fill=1)
# make text go straight up
c.rotate(90)
# change color
c.setFillColorRGB(0,0,0.77)
# say hello (note after rotate the y coord needs to be negative!)
c.drawString(0.3*inch, -inch, "Hello World")

# Now Save and close PDF
c.showPage()
c.save()
```

```python
from reportlab.pdfgen import canvas

from reportlab.lib.units import inch
from reportlab.lib.colors import lightblue, white, blue, black, gray, red
from reportlab.lib.units import cm

# Create the Canvas
c = canvas.Canvas('type_1.pdf')

# Set meta Info
c.setAuthor('Website Preneur')
c.setTitle('Type 1 - Payment Reciept')
c.setSubject('Reciept for payment made')

# ---------------- Start drawing on canvas ---------------

c.setFillColor(lightblue)
c.rect(0,0,2.5 * inch,11.8 * inch, fill=True, stroke=False)

c.setFillColor(blue)
c.rect(0,10.3 * inch,2.5 * inch,1.5 * inch, fill=True, stroke=False)

c.setStrokeColor(white)
c.setFillColor(gray)
c.circle(1.25*inch, 9.3*inch, 0.75*inch, fill=1, stroke =1)

c.setFillColorCMYK(0,0,0,0)
c.setFont("Helvetica-Bold", 20)
c.drawString(0.6 * inch, 10.5 * inch, "RECIEPT")

c.setFillColorCMYK(0,0,0,0)
c.setFont("Helvetica-Bold", 20)
c.drawCentredString(1.25*inch, 9.2*inch, "LOGO")

text_1 = [
    'Your Company Name',
    'Street Address',
    'City, State, Zip/Post Code',
    'Phone Number',
    'E-Mail',
    'Website'
]
cord = 7.5
x = 0.5 * inch
y = 7.5 * inch

c.setFillColorCMYK(0.8,0.7,0.7,1)

for line in text_1:
    c.setFont("Helvetica", 10)
    c.drawString(x,y,line)
    cord -= 0.3
    y = cord * inch

c.setFont("Helvetica-Bold", 10)
c.drawCentredString(7.0 * inch, 10.6 * inch,'DATE')    
    
c.setStrokeColor(black)
c.line(6.5 * inch, 10.5* inch, 7.5 * inch, 10.5 * inch)

c.drawCentredString(7.0 * inch, 9.6 * inch,'RECIEPT NO')   
c.setStrokeColor(black)
c.line(6.5 * inch, 9.5* inch, 7.5 * inch, 9.5 * inch)


c.drawString(3.0 * inch, 8.5 * inch,'BILLED TO')  

c.setStrokeColor(black)
c.line(3.0 * inch, 8.4* inch, 5.0 * inch, 8.4 * inch)

c.drawString(5.5 * inch, 8.5 * inch,'SHIPED TO')  

c.setStrokeColor(black)
c.line(5.5 * inch, 8.4* inch, 7.5 * inch, 8.4 * inch)

  

cord = 7.5
x = 3.0 * inch
y = 8.1 * inch

c.setFillColorCMYK(0.8,0.7,0.7,1)

for line in text_1[:-1]:
    c.setFont("Helvetica", 10)
    c.drawString(x,y,line)
    y = y - 0.2*inch

x = 5.5 * inch
y = 8.1 * inch

for line in text_1[:-2]:
    c.setFont("Helvetica", 10)
    c.drawString(x,y,line)
    y = y - 0.2*inch 



    


c.setStrokeColor(black)
c.line(3.0 * inch, 1 * inch, 5.0 * inch, 1 * inch)  
c.drawString(3.0 * inch, 0.8 * inch,'Company\'s Signature')

c.setStrokeColor(black)
c.line(5.5 * inch, 1 * inch, 7.5 * inch, 1 * inch)  
c.drawString(5.5 * inch, 0.8 * inch,'Customer\'s Signature')

# -------------------- Show Page --------------
c.showPage()
c.save()
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.lib.colors import magenta, red
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors


canvas = canvas.Canvas('texts.pdf')
canvas.setFont("Times-Roman", 20)
canvas.setFillColor(red)
canvas.drawCentredString(2.75*inch, 2.5*inch, ".Font size Centered examples")
canvas.drawString(2.75*inch, 2.0*inch, ".Font size examples - Draw String")
canvas.drawRightString(2.75*inch, 1.5*inch, ".Font size examples - Draw Right String")
canvas.setFillColor(magenta)
# ------------------------
lyrics = [
    'This is line 1',
    'This is line 2',
    'This is line 3',
    'This is line 4',
    'This is line 5',
    'This is line 6'
]
size = 7
y = 10.3*inch
x = 1.3*inch
for line in lyrics:
    canvas.setFont("Helvetica", size)
    canvas.drawRightString(x,y,"%s points: " % size)
    canvas.drawString(x,y, line)
    y = y-size*1.2
    #size = size+1.5

canvas.line(50, 380, 560, 380)
canvas.setLineWidth(5)
p = canvas.beginPath()
p.moveTo(inch,2 * inch)
p.lineTo(7 * inch,6 * inch)
p.close
canvas.drawPath(p)



elements = []

data= [
    ['00', '01', '02', '03', '04'],
    ['10', '11', '12', '13', '14'],
    ['20', '21', '22', '23', '24'],
    ['30', '31', '32', '33', '34']
]

t=Table(data)
t.setStyle(
    TableStyle(
        [
            ('BACKGROUND',(1,1),(-2,-2),colors.green),
            ('TEXTCOLOR',(0,0),(1,-1),colors.red)
        ]
    )
)
width = 400
height = 100
x = 100
y = 800

t.wrapOn(p, width, height)
t.drawOn(p, x, y)


# ------------------------

canvas.showPage()
canvas.save()
```
